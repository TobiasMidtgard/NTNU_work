```python
A = np.zeros((4, 4))

x_values = np.array([-1, 1, 2, 3])

for i in range(len(x_values)):
    A[i] = np.array([x_values[i]**3, x_values[i]**2, x_values[i], 1])

y_values = np.array([3, 2, -1, 4])

consts = np.linalg.solve(A, y_values)
a = consts[0]
b = consts[1]
c = consts[2]
d = consts[3]
```

b)

```python
from matplotlib import pyplot as plt
```

```python
N = 1000
x = np.linspace(-1.5, 3.5, N, endpoint=True)
y = a * (x ** 3) + b * (x ** 2) + c * x + d

plt.plot(x, y, color='black', lw=2, label='f(x)')
plt.scatter(x_values, y_values, color='red', marker='o', s=50, zorder=2.5, label='points')
# plt.plot(x_2, y_2, 'ro', zorder=2.5)

plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Oppgave 2', fontsize=16)
plt.text(-1, 7, r'$f(x)=ax^{3}+bx^{2}+cx+d$', fontsize=12)

plt.legend()

plt.grid()
plt.show()
```
